```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
playertotal = len(purchase_data["SN"].value_counts())
playertotal
```

```python
table_pt = pd.DataFrame({'Total Players':[playertotal]})
table_pt
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_of_unique_items = len((purchase_data["Item ID"]).unique())
number_of_unique_items
```

```python
average_price = (purchase_data["Price"]).mean()
average_price
```

```python
purchase_count = (purchase_data["Purchase ID"]).count()
purchase_count
```

```python
total_spent = (purchase_data["Price"]).sum()
total_spent
```

```python
purchase_analysis_df = pd.DataFrame({
    "Unique Item Count":[number_of_unique_items],
    "Average Price":[average_price],
    "Purchase Count":[purchase_count],
    "Total Earnings":[total_spent]
})
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ages = [0,9,14,19,24,29,34,39,1000]
agegroups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"],ages, labels=agegroups)
purchase_data.head()
```

```python
age_df = purchase_data.groupby("Age Group")
age_df.head()
```

```python
total_by_age = age_df["SN"].nunique()
total_by_age
```

```python
percentage_by_age = (total_by_age/playertotal)*100
percentage_by_age
```

```python
age_group_demo_df = pd.DataFrame({
    "Percentage of Players":percentage_by_age,
    "Total Count":total_by_age
})

age_group_demo_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = purchase_data.groupby("SN")
```

```python
purchase_count_by_spender = top_spender["Purchase ID"].count()
purchase_count_by_spender
```

```python
avg_price_per_spender = top_spender["Price"].mean()
avg_price_per_spender
```

```python
top_spender_df = pd.DataFrame({
    "Purchase Count":purchase_count_by_spender,
    "Average Purchase Price":avg_price_per_spender,
    "Total Purchase Value":total_purchase_per_spender
})
top_spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = purchase_data[["Item ID","Item Name", "Price"]]
item_df.head()
```

```python
item_id_df = item_df.groupby(["Item ID","Item Name"])
item_id_df.head()
```

```python
item_purchase_count = item_id_df["Price"].count()
item_purchase_count
```

```python
item_purchase_value = item_id_df["Price"].sum()
item_purchase_value
```

```python
price_by_item = item_purchase_value/item_purchase_count
price_by_item
```

```python
most_pop_item = pd.DataFrame({
    "Purchase Count":item_purchase_count,
    "Item Price":price_by_item,
    "Total Purchase Value":item_purchase_value
})

most_pop_item
```

```python
most_pop_item = most_pop_item.sort_values(["Purchase Count"], ascending=False)
most_pop_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profit_item = most_pop_item.sort_values(["Total Purchase Value"], ascending=False)
most_profit_item.head()
```
